```python
#Question 1
import numpy as np
import scipy as sp
from scipy.linalg import qr
from numpy.linalg import inv
from sklearn import preprocessing
from sympy import *


# Projects vector 'b' onto vector 'a'
def proj(b, a):
    return (np.dot(a, b) / np.linalg.norm(a) ** 2) * a

M = Matrix([[1, 2, 3, 4], [0, 1, 0, 1], [1, 4, 3, 6], [2, 11, 6, 15]])
#Part 1: produce vector that's in S
#This produces a vector that is a linear combination of the subspace S which means it is in S
v1 = np.array([1, 2, 3, 4])
v2 = np.array([0, 1, 0, 1])
v3 = np.array([1, 4, 3, 6])
v4 = np.array([2, 11, 6, 15])

num1 = np.random.randint(10,size = 1)
num2 = np.random.randint(10,size = 1)
num3 = np.random.randint(10,size = 1)
num4 = np.random.randint(10,size = 1)

vectorInS = (v1*num1) + (v2*num2) + (v3*num3) + (v4*num4)
print("Vector in S: " + str(vectorInS))

##produce vector not in S
#not a linear combination of the subspace S (doesn't span S) = {v1, v2, v3, v4}
vecNotInS = np.random.rand(1, 4)[0]
print("Vector not in S: " + str(vecNotInS))

##create a vector perpendicular to S
##project a random vector onto one of the vectors in S
rref = M.rref() #basis of S
v1 = rref[0].row(0)
v2 = rref[0].row(1)
a = np.array(v1).astype(np.float64)
b = np.random.rand(1, 4)[0]
b = np.array(b).astype(np.float64)
a = np.squeeze(np.asarray(a))
b = np.squeeze(np.asarray(b))

parallelVector = (np.dot(a, b)) / (np.linalg.norm(a) ** 2) * a
print("New parallel vector to S: " + str(parallelVector))

##To check if a vector b is in S, put the vector b in an augmented matrix with S
##Row reduce to RREF and if there are any free variables, then b is not in S


#Part 2: find dimension of S
dimension = len(rref[1])
print("dimension of S: " + str(dimension))


#Part 3: find orthonormal basis for subspace S
basis = Matrix([rref[0].row(0), rref[0].row(1)])

#use Gram-Schmidt to find orthonormal basis (v1, v2)
basisList = np.array(rref[0].tolist())

basisOfS = [v1, v2]
print("Orthonormal basis of S: " + str(GramSchmidt(basisOfS, True)))


#part 4, solve the optimization problem
z = Matrix([[1,0,0,0]])
z = z.T
A = basis.T

#Use formula: A(A^TA)^-1*A^T*z

closestPoint = A*((A.T*A)**-1)*A.T*z
print("Solution to optimization problem: " + str(closestPoint))
```

```python
#Question 2
from bs4 import BeautifulSoup as bs
import requests
import urllib

def getPDFName(link):
    namePDF = str(link)
    last_index = namePDF.rfind('/')
    name = namePDF[last_index + 1:]
    
    return name
    
def writeFile(name, link):
     r = requests.get(link)
     with open(name,'wb') as f:
        f.write(r.content)

def scrape():
    req = requests.get("http://proceedings.mlr.press/v70/")
    html = req.content
    soup = bs(html, "lxml")
    links = soup.find_all("div", attrs = {"class": "paper"})
    for link in links:
        papers = link.find_all('a')
        try:
            mainPDF_link = papers[1]['href']
            name = getPDFName(mainPDF_link)
            writeFile(name, mainPDF_link)
        except IndexError:
            
        try:
            supplPDF_link = papers[2]['href']
            name = getPDFName(supplPDF_link)
            writeFile(name, supplPDF_link)
        except IndexError:
           

scrape()

```

```python
#2b 
from os import listdir
from os.path import isfile, join
from collections import Counter
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from io import StringIO

path = '/Users/pratyushsingh/Data Science Lab EE379K/Lab3/'
files = [f for f in listdir(path) if isfile(join(path, f)) and f.find(".pdf") != -1]
wordCount = {}
words = []

def processWords(words):
    counter = Counter()
    for idx, word in enumerate(words):
        word = word.rstrip()
        word = word.lstrip()
        counter[word] += 1
    
    return counter
    
        
for file in files:
    rsrcmgr = PDFResourceManager()
    retstr = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
    fp = open(file, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    password = ""
    maxpages = 0
    caching = True
    pagenos = set()
    
    try:
        for page in PDFPage.get_pages(fp, pagenos, maxpages=maxpages,caching=caching, check_extractable=True):
            interpreter.process_page(page)

        text = retstr.getvalue()
        text = text.replace('\n', ' ')
        text = text.split(' ')

        words = words + text
        print("Processed: " + file + " succesfully")
    
    except:
        print("Error in file: " + file)
        
    fp.close()
    device.close()
    retstr.close()
    
```

```python
#2 continued
from nltk.corpus import words as english_dict
from nltk.probability import FreqDist
from nltk.stem.porter import PorterStemmer

words = [word for word in words if word.isalpha()] #remove punctuation
words = [word.lower() for word in words] #make lowercase
english = set(english_dict.words())
tokens = []

def remove_words():
    for idx, word in enumerate(words):
        if(word in english):
            tokens.append(word)

remove_words() #removing words that are not in the english dictionary

#taking the stems of words 
porter = PorterStemmer()
stemmed = [porter.stem(word) for word in tokens] #stemming the words
freq_distribution_stemmed = FreqDist(stemmed) #frequency distribution of the stems
most_common_stems = freq_distribution_stemmed.most_common(10) #most common stems

#ten most common words without removing the stop words
freq_distribution_words = FreqDist(tokens)
most_common_words = freq_distribution_words.most_common(10)

print("The ten most common stems are: " + str(most_common_stems))
print()
print()
print("The most common words are: " + str(most_common_words))
print()
print()

#lets remove the stop words 
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words_removed = [w for w in tokens if not w in stop_words]
freq_distribution_ = FreqDist(words_removed)
print("The most common words after the stop-words have been removed: " + str(freq_distribution_.most_common(10)))
```

```python
#calculating entropy
from math import log

freq = {}
wordsSeen = set()
def calculate_relative_frequency():
    for token in tokens:
        if token in wordsSeen:
            continue
        else:
            freq[token] = freq_distribution_words[token]/len(tokens)
        
def entropy():
    entropy = 0
    for word, frequency in freq.items():
        p_i = frequency
        entropy_of_term = p_i * log(p_i, 2)
        entropy = entropy + entropy_of_term
    
    return entropy * -1

calculate_relative_frequency()
entropy = entropy()
print(entropy)
```

```python
#synthesizing a paragraph based on the marginal distributions
import numpy as np

keys = list(freq.keys()) #words
frequency_of_keys = list(freq.values()) #individual frequencies

synth_random_paragraph = np.random.choice(keys, 100, p=frequency_of_keys) #print 1000 words
for w in synth_random_paragraph:
    print(w + " ", sep=' ', end='', flush=True)

print()
print()

random_freq = FreqDist(synth_random_paragraph) 
print("The ten most common words from random paragraph: " + str(random_freq.most_common(10)))

```

```python
#3
import pandas as pd
import seaborn as sn
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import skew
from scipy.stats.stats import pearsonr

training = pd.read_csv('Housing Data/train.csv')
testing = pd.read_csv('Housing Data/test.csv')

%config InlineBackend.figure_format = 'png' #set 'png' here when working on notebook
%matplotlib inline

training.head()
```

```python
all_data = pd.concat((training.loc[:,'MSSubClass':'SaleCondition'],
                      testing.loc[:,'MSSubClass':'SaleCondition']))

matplotlib.rcParams['figure.figsize'] = (12.0, 6.0)

#showing the difference after normalization vs pre-normalization
prices = pd.DataFrame({"price":training["SalePrice"], "log(price + 1)":np.log1p(training["SalePrice"])})
prices.hist()
```

```python
#normalize the data
training['SalePrice'] = np.log1p(training['SalePrice'])

#transform the numeric features
numeric_feats = all_data.dtypes[all_data.dtypes != "object"].index

skewed_feats = training[numeric_feats].apply(lambda x: skew(x.dropna())) #compute skewness of each feature
skewed_feats = skewed_feats[skewed_feats > 0.75] #features with skew greater than a value of 0.75
skewed_feats = skewed_feats.index #get the index of these feautures

all_data[skewed_feats] = np.log1p(all_data[skewed_feats]) #all skewed features are normalized
all_data = pd.get_dummies(all_data) # turn the categorical values into indicator variables
all_data = all_data.fillna(all_data.mean())

all_data.head()
```

```python
from sklearn.linear_model import Ridge, RidgeCV, ElasticNet, LassoCV, LassoLarsCV
from sklearn.model_selection import cross_val_score

#split the training data into train and testing
x_train = all_data[:training.shape[0]]
x_test = all_data[training.shape[0]:]
y = training.SalePrice

def rmse_cv(model):
    rmse= np.sqrt(-cross_val_score(model, x_train, y, scoring="neg_mean_squared_error", cv = 10))
    return(rmse)

alphas = [0.05, 0.1, 0.3, 1, 3, 5, 10, 15, 30, 50, 75] #making a list of alphas to choose from
model_ridge = Ridge() #using the ridge model
cv_ridge = [rmse_cv(Ridge(alpha = alpha)).mean() for alpha in alphas]

cv_ridge = pd.Series(cv_ridge, index = alphas)
cv_ridge.plot()
plt.xlabel("alpha values")
plt.ylabel("rmse values")
plt.show()

print(cv_ridge.min())
```

```python
#3 continued
#making a model using an alpha of 10
ridge = Ridge(10)
ridge_model = ridge.fit(x_train, y)
ridge_preds = np.expm1(ridge_model.predict(x_test))

solution = pd.DataFrame({"id":testing.Id, "SalePrice":ridge_preds})
solution.to_csv("ridge_sol_new.csv", index = False)
```

```python
#3 continued... trying to improve the model
#using the lassos regression model
lassos = LassoCV(alphas = [1, 0.1, 0.001, 0.0005], max_iter=1e5).fit(x_train, y)
coef = pd.Series(lassos.coef_, index=x_train.columns)
print("Lasso picked " + str(sum(coef != 0)) + " variables and eliminated the other " +  str(sum(coef == 0)) + " variables")
```

```python
#3 continued
#exploring the coeffiecients that are most important 
important_coef = pd.concat([coef.sort_values().tail(10),
                     coef.sort_values().head(10)])
important_coef.plot(kind = "barh")
plt.title("Coefficients in the Lasso Model")
```

```python
#making predictions using lassos
prediction = np.expm1(lassos.predict(x_test))
solution = pd.DataFrame({"id":testing.Id, "SalePrice":prediction})
solution.to_csv("lassos_solution.csv", index = False)
```
